```python
import pandas as pd
```

```python
from pathlib import Path

folder_path = Path('busStop/ขดร.1')

def get_file_names(folder_path):
    return [f.name for f in folder_path.iterdir() if f.is_file()]

print(get_file_names(folder_path))
```

```python
columns = ['Station name', 'lat', 'lng']

# Create empty DataFrame
df = pd.DataFrame(columns=columns)

print(df)
```

```python
def getbus_stop(file_name, folder_index):
    return pd.read_excel('busStop/ขดร.'+folder_index+'/'+file_name, skiprows=4)
    
    
```

```python
for i in range(1, 9):
    files = get_file_names(Path('busStop/ขดร.'+str(i)))
    for file in files:
        df_temp = getbus_stop(file, str(i))
        # print(df_temp.head(1))
        df_temp = df_temp.rename(columns={'ชื่อสถานี': 'Station name', 'ละติจูด': 'lat', 'ลองจิจูด': 'lng'})
        df_temp = df_temp[['Station name', 'lat', 'lng']]
        df = pd.concat([df, df_temp], ignore_index=True).drop_duplicates(subset=['Station name'])
```

```python
df.shape, df.head()
df[['lat','lng']].value_counts()
```

```python
df.drop_duplicates(subset=['lat', 'lng'], keep='first', inplace=True)
```

```python
df.rename(columns={'lat': 'lat', 'lng': 'lon', 'Station name': 'name'}, inplace=True)
df.to_csv('data/busStop.csv', index=False, encoding='utf-8-sig')
```
