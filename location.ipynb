```python
import pandas as pd
```

```python
df = pd.read_csv('data/location.csv')
stations = pd.read_csv('data/bangkok_bts_stations.csv')
```

```python
df['name'].isna().sum()
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['name'], inplace=True)
df.describe()
```

```python
df['type'].value_counts()
```

```python
df.head(20)
```

```python
stations.head()
```

```python
stations['Coordinates'] = stations['Coordinates'].str.replace(',','').str.replace('°N', ',').str.replace('°E', '')

stations.head()
```

```python
stations[['lat', 'lon']] = stations['Coordinates'].str.split(",", expand=True).astype(float)

stations.head()
```

```python
stations.drop(columns=['Coordinates', 'Lines'], inplace=True)
stations.head()
```

```python
stations['type'] = 'RapidTransitStation'
```

```python
locations =pd.concat([df,stations.rename(columns={'Station Name (Thai)':'name', 'Coordinates':'coords'})], ignore_index=True)
```

```python
locations.head()
```

```python

locations.tail()
```

```python
locations.to_csv('data/bangkok_locations.csv', index=False)
```
